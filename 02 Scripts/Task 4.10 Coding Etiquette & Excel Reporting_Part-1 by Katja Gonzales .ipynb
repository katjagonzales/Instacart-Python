# 04. Comparing Customer Behaviour in Different Geographic Areas

## Q3. Create a regional segmentation of the data. You’ll need to create a “Region” column based on the “State” column from your customers data set.



### a) Use the region information in this Wikipedia article to create your column (you only need to create regions, not divisions).

```python
# Check the frequencies of the 'state' column

df_ords_prods_custm['state'].value_counts(dropna = False)
```

```python
# Group the US states by region in a list

Northeast = ['Maine', 'New Hampshire', 'Vermont', 'Massachusetts', 'Rhode Island', 'Connecticut', 'New York', 'Pennsylvania', 'New Jersey']
```

```python
Midwest = ['Wisconsin', 'Michigan', 'Illinois', 'Indiana', 'Ohio', 'North Dakota', 'South Dakota', 'Nebraska', 'Kansas', 'Minnesota', 'Iowa', 'Missouri']
```

```python
South = ['Delaware', 'Maryland', 'District of Columbia', 'Virginia', 'West Virginia', 'North Carolina', 'South Carolina', 'Georgia', 'Florida', 'Kentucky',
'Tennessee', 'Mississippi', 'Alabama', 'Oklahoma', 'Texas', 'Arkansas', 'Louisiana']
```

```python
West = ['Idaho', 'Montana', 'Wyoming', 'Nevada', 'Utah', 'Colorado', 'Arizona', 'New Mexico', 'Alaska', 'Washington', 'Oregon', 'California', 'Hawaii']
```

```python
# Assign the region list values to a new 'region' column

df_ords_prods_custm.loc[df_ords_prods_custm['state'].isin(Northeast), 'region'] = 'Northeast'
```

```python
df_ords_prods_custm.loc[df_ords_prods_custm['state'].isin(Midwest), 'region'] = 'Midwest'
```

```python
df_ords_prods_custm.loc[df_ords_prods_custm['state'].isin(South), 'region'] = 'South'
```

```python
df_ords_prods_custm.loc[df_ords_prods_custm['state'].isin(West), 'region'] = 'West'
```

```python
# Check the new column 'region'

df_ords_prods_custm.head()
```

```python
#  Check the frequencies for the new 'region' column

df_ords_prods_custm['region'].value_counts(dropna = False)
```

### b) Determine whether there’s a difference in spending habits between the different U.S. regions. (Hint: You can do this by crossing the variable you just created with the spending flag.)

```python
# Create a spending_habits crosstab

spending_habits_crosstab = pd.crosstab(df_ords_prods_custm['region'], df_ords_prods_custm['spending_flag'], dropna = False)
```

```python
# Check the output of the crosstab

spending_habits_crosstab
```

#### Spending habits vary between the different regions, but in general, there is a higher number Low spender customers in all the areas. Nevertheless the South presents the higher values in both cases.

# 05. Excluding low-activity customers

## Q4. The Instacart CFO isn’t interested in customers who don’t generate much revenue for the app. Create an exclusion flag for low-activity customers (customers with less than 5 orders) and exclude them from the data. Make sure you export this sample.

```python
# Create an exclusion flag 

df_ords_prods_custm.loc[df_ords_prods_custm['max_order'] < 5, 'exclusion_flag'] = 'low-activity'
```

```python
df_ords_prods_custm.loc[df_ords_prods_custm['max_order'] >= 5, 'exclusion_flag'] = 'high-activity'
```

```python
# Check the frequencies of the exclusion flag

df_ords_prods_custm['exclusion_flag'].value_counts()
```

```python
# Check the results

crosstab_exclusion_flag = pd.crosstab(df_ords_prods_custm['max_order'], df_ords_prods_custm['exclusion_flag'], dropna = False)
```

```python
# Check the output of the crosstab

crosstab_exclusion_flag
```

```python
# Create a subset which excludes low-activity customers

df_ords_prods_custm2 = df_ords_prods_custm.loc[df_ords_prods_custm['exclusion_flag'] == 'high-activity']
```

```python
# Export the new dataframe 

df_ords_prods_custm2.to_pickle(os.path.join(path, '02 Data', 'Prepared Data', 'ords_prods_high_activity_customers.pkl'))
```

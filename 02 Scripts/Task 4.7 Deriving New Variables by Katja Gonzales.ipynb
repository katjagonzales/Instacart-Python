# 03. If-Statements with User-Defined Functions

```python
# Define a function

def price_label(row):
    
    if row['prices'] <= 5:
        return 'Low-range product'
    elif (row['prices'] > 5) and (row['prices'] <= 15):
         return 'Mid-range product'
    elif row['prices'] > 15:
         return 'High-range product'
    else: return np.nan
```

```python
# Apply the function

df['price_range'] = df.apply(price_label, axis=1)
```

```python
# Check the frequency of the 'price range' column

df['price_range'].value_counts(dropna = False)
```

```python
# Find the most expensive product

df['prices'].max()
```

### 2. Suppose your clients have changed their minds about the labels you created in your busiest_day column. Now, they want Busiest day to become Busiest days (plural). This label should correspond with the two busiest days of the week as opposed to the single busiest day. At the same time, theyd also like to know the two slowest days. Create a new column for this using a suitable method.

```python
# Check the frequency of the 'order_day_of_week' column

df_ords_prods_merged['order_day_of_week'].value_counts(dropna = False)
```

#### The two busiest days are Saturday(0) and Sunday(1), while the two slowest days are Tuesday(3) and Wednesday(4).

```python
# Create a for-loop that appends to the 'result2' list the busyness of the days based on values in the 'order_day_of_week' column

result2 = []

for value in df_ords_prods_merged['order_day_of_week']:
    if value == 0 or value == 1:
        result2.append('Busiest days')
    elif value == 3 or value == 4:
        result2.append('Slowest days')
    else:
        result2.append('Regularly busy days')
        
```

```python
# Create a new column in df_ords_prods_merged called 'busiest_days'

df_ords_prods_merged['busiest_days'] = result2
```

## 4. When too many users make Instacart orders at the same time, the app freezes. The senior technical officer at Instacart wants you to identify the busiest hours of the day. Rather than by hour, they want periods of time labeled Most orders, Average orders, and Fewest orders. Create a new column containing these labels called busiest_period_of_day.

```python
# Check the frequency of the values in the 'order_hour_of_day' column

df_ords_prods_merged['order_hour_of_day'].value_counts(dropna = False)
```

#### By splitting the 24 hours of the day into three parts,  'Most orders' are at hours 10, 11, 14, 15, 13, 12, 16, 9 - meaning from 9am to 4pm, 'Average orders' are at hours 17, 8, 18, 19, 20, 7, 21, 22 - meaning from 7am to 8am and from 5pm to 10pm, and 'Fewest orders' are at hour 23, 6, 0, 1, 5, 2, 4, 3 - meaning from 11pm to 6am.

```python
# Create a for-loop that appends to the 'hours' list the busyness of the hours based on values in the 'order_hour_of_day' column

hours = []

for value in df_ords_prods_merged['order_hour_of_day']:
    if value in [10, 11, 14, 15, 13, 12, 16, 9]:
        hours.append('Most orders')
    elif value in [23, 6, 0, 1, 5, 2, 4, 3]:
        hours.append('Fewest orders')
    else:
        hours.append('Average orders')
```

```python
# Create the 'busiest_period_of_day' column

df_ords_prods_merged['busiest_period_of_day'] = hours
```

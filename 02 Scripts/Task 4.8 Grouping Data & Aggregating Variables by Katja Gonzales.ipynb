## 6. The team now wants to target different types of spenders in their marketing campaigns. This can be achieved by looking at the prices of the items people are buying. Create a spending flag for each user based on the average price across all their orders using the following criteria:

### a) If the mean of the prices of products purchased by a user is lower than 10, then flag them as a “Low spender"

### b) If the mean of the prices of products purchased by a user is higher than or equal to 10, then flag them as a “High spender.”

```python
# Create a "spending flag" column 'average_price' using groupby() and transform()

df_ords_prods_merge['average_price'] = df_ords_prods_merge.groupby(['user_id'])['prices'].transform(np.mean)
```

```python
# Check the output

df_ords_prods_merge.head()
```

```python
# a) Create the "spending flag" flag using the loc function

df_ords_prods_merge.loc[df_ords_prods_merge['average_price'] < 10, 'spending_flag'] = 'Low spender'
```

```python
# b) Create the "spending flag" flag using the loc function

df_ords_prods_merge.loc[df_ords_prods_merge['average_price'] >= 10, 'spending_flag'] = 'High spender'
```

```python
# Check the output

df_ords_prods_merge.head()
```

```python
# Check the frequency of the 'spending_flag' column

df_ords_prods_merge['spending_flag'].value_counts(dropna = False)
```

```python
# Check the output of just the 'user_id', 'average_price' and 'spending_flag' column

df_ords_prods_merge[['user_id', 'average_price', 'spending_flag']].tail(20)
```

## 7. In order to send relevant notifications to users within the app (for instance, asking users if they want to buy the same item again), the Instacart team wants you to determine frequent versus non-frequent customers. Create an order frequency flag that marks the regularity of a user’s ordering behavior according to the median in the “days_since_prior_order” column. The criteria for the flag should be as follows:

### a) If the median of “days_since_prior_order” is higher than 20, then the customer should be labeled a “Non-frequent customer.”

### b) If the median is higher than 10 and lower than or equal to 20, then the customer should be labeled a “Regular customer.”

### c) If the median is lower than or equal to 10, then the customer should be labeled a “Frequent customer.”

```python
# Create a 'order_frequency' column using groupby() and transform()

df_ords_prods_merge['order_frequency'] = df_ords_prods_merge.groupby(['user_id'])['days_since_prior_order'].transform(np.median)
```

```python
# Check the output

df_ords_prods_merge.head()
```

```python
# a) Create the "frequency" flag using the loc function

df_ords_prods_merge.loc[df_ords_prods_merge['order_frequency'] > 20, 'frequency_flag'] = 'Non-frequent customer'
```

```python
# b) Create the "frequency" flag using the loc function

df_ords_prods_merge.loc[(df_ords_prods_merge['order_frequency'] > 10) & (df_ords_prods_merge['order_frequency'] <= 20), 'frequency_flag'] = 'Regular customer'
```

```python
# c) Create the "frequency" flag using the loc function

df_ords_prods_merge.loc[df_ords_prods_merge['order_frequency'] <= 10, 'frequency_flag'] = 'Frequent customer'
```

```python
# Check the frequency of the 'frequency_flag' column

df_ords_prods_merge['frequency_flag'].value_counts(dropna = False)
```

```python
# Check for missing values

df_ords_prods_merge.isnull().sum()
```

# 03. Create Visualizations

### 2. If you haven’t done so already, follow the instructions in the Exercise to create a bar chart, a histogram, a scatterplot, and a line chart, using a sample subset for the line chart.

### Bar Chart

```python
# Create bar charts

df_ords_prods_custm['order_day_of_week'].value_counts().plot.bar()
plt.show()
```

```python
# Sort according to the index order

bar_ords_week = df_ords_prods_custm['order_day_of_week'].value_counts().sort_index().plot.bar()
plt.xlabel("")
plt.ylabel("Orders placed(in millions)")
plt.title ("Instacart Orders Placed by Day of the Week")
plt.xticks([0,1,2,3,4,5,6],['Saturday','Sunday','Monday','Tuesday','Wednesday','Thursday','Friday'])
plt.show()
```

#### Most orders are placed at the weekend (Saturday and Sunday).

```python
# Print frequencies

df_ords_prods_custm['order_day_of_week'].value_counts()
```

### Histogram

```python
# Create a histogram for 'prices'

hist_prices = df_ords_prods_custm['prices'].plot.hist(bins = 70)
plt.xlabel("Prices (USD)")
plt.ylabel("Frequency(in millions)")
plt.title ("Frequency of Prices for Instacart Products")
plt.show()
```

### Scatterplot

```python
# Create a scatterplot

scatterplot_prices = sns.scatterplot(x = 'prices', y = 'prices',data = df_ords_prods_custm)
```

### Sampling Data

```python
# Create a sample of the dataframe: split the dataframe into two dataframes at a 70/30 ratio

np.random.seed(4)
dev = np.random.rand(len(df_ords_prods_custm)) <= 0.7
```

```python
# Store 70% of the sample in the dataframe big

big = df_ords_prods_custm[dev]
```

```python
# Store 30% of the sample in the dataframe small

small = df_ords_prods_custm[~dev]
```

```python
# Check the number of rows in the original dataframe

len(df_ords_prods_custm)
```

```python
# Check if the sum of the samples is consistent with the original number of rows

len(big) + len(small)
```

```python
# Create a subset data frame for the line chart

df_small = small[['order_day_of_week','prices']]
```

### Line Chart

```python
# Create the line chart

line_chart_week_prices = sns.lineplot(data = df_small, x = 'order_day_of_week', y = 'prices')
plt.xlabel(" order day of the week")
plt.ylabel("Prices of products purchased (USD)")
plt.title ("Prices of Instacart products Purchased by Day of the Week")
plt.show()
```

### 3. You need to provide the Instacart senior stakeholders with descriptive findings about sales. Create a histogram of the “order_hour_of_day” column.



### In a markdown cell beneath your histogram, describe what the histogram demonstrates.


```python
# Create a histogram for orders placed by hour

hist_order_hour_of_day = df_ords_prods_custm['order_hour_of_day'].plot.hist(bins = 24)
plt.xlabel("Hour of  the day")
plt.ylabel("Frequency (in millions)")
plt.title ("Frequency of Instacart Orders per Hour of the Day")
plt.show()
```

#### In the histogram you can see that most orders are placed in the middle of the day between 10:00 and 17:00. 

### 4. The marketing team is curious about the distribution of orders among customers in terms of loyalty. Create a bar chart from the “loyalty_flag” column.

```python
# Create the bar chart 

bar_loyalty_customers = df_ords_prods_custm['loyalty_flag'].value_counts().plot.bar(rot=0, color = ['darkblue', 'blue', 'lightblue'])
plt.xticks([0,1,2],['Regular customer\n(10-40 orders)','Loyal customer\n(>40 orders)','New customer\n(<10 orders)'])
plt.ylabel("number of customers (in millions)")
plt.title ("Loyalty status")
plt.show()
```

#### Most customers are regular customers, followed by loyal customers.

### 7. You’ll also need to explore whether there’s a connection between age and spending power (income). To visualize this relationship, create a scatterplot using the sns.scatterplot() function.



### In a markdown cell beneath your scatterplot, describe what the scatterplot shows.


```python
# Create a scatterplot

scatterplot_age_income = sns.scatterplot(x = 'age', y = 'income',data = df_ords_prods_custm)
plt.xlabel("Age (in years)")
plt.ylabel("Income (USD)")
plt.title ("Income According to Age")
```

#### The scatterplot above shows age at the x axis and income at the y axis. There is a positive correlation beteen age and income. We can see that the higher the age is, the higher the income is. The majority of customers between 40 and 80 years old have an income above 200000, while an income over 400000 is only present in the above 40 years age group.
